```python
df_compare=df[["satisfaction_level","average_montly_hours","promotion_last_5years","salary"]]
```

```python
df_compare
```

```python
salary_dummies=pd.get_dummies(df.salary)
```

```python
salary_dummies
```

```python
merged=pd.concat([df_compare,salary_dummies],axis="columns")
```

```python
merged
```

```python
merged.drop(["high","salary"],axis="columns",inplace=True)
```

```python
merged
```

```python
merged=pd.concat([merged,df.left],axis="columns")
```

```python
merged
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
x=merged[["satisfaction_level",'average_montly_hours','promotion_last_5years','low',]]
y=merged.left
```

```python
from sklearn.model_selection import train_test_split

X_train,x_test,Y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model.fit(X_train,Y_train)
```

```python
ans=model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test,y_test)
```
